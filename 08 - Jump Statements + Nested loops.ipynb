#### Question
- Write a program that continuously asks the user to provide an input number.
- Print the square of that input number.
- The program should stop if the user enters 5.
- In case user enters any multiple of 4, skip it.

````python
while True:
    x = int(input())
    
    if x == 5:
        break
        
    if x % 4 == 0:
        continue
    
    print(x * x)
````

##### **Question**: Write a program to print a `N x M` matrix of `*`

````python
# N = rows
# M = cols
````

````python
N = int(input())
M = int(input())

for i in range(N):
    for j in range(M):
        print('*', end='')
    print()
````

## Stair Pattern

````python
N = 5

for i in range(1, N+1):
    # For every row, we want to print i stars
    for j in range(i):
        print('*', end='')
        
    print() # the default value of end='\n' which will move the output to the new line
````

````python
# Run a loop that prints numbers from 1 to 10
N = 10

for i in range(1, N+1):
    print(i, end=' ')
````

## Reverse Stair Pattern

````python
N = 10

for i in range(1, N+1):
    # For every row, print (N-i+1) stars
    for j in range(N-i+1):
        print('*', end='')
    print()
````

# GCD
Given two integer inputs, write a program to calculate GCD of these two numbers.


![](https://www.scaler.com/topics/images/gcd.webp)

````python
A = 15
B = 10

for i in range(min(A, B), 0, -1):
    # We are looking for common factors of A & B
    if A % i == 0 and B % i == 0:
        # i is a common factor of A & B
        print(i)
        break
````
